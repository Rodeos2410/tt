```python
class Автомобиль:
    def __init__(self, марка, модель, год_выпуска, пробег=0):
        self.марка = марка
        self.модель = модель
        self.год_выпуска = год_выпуска
        self.пробег = пробег

    def информация(self):
        """Выводит всю информацию об автомобиле."""
        print(f"Информация об автомобиле")
        print(f"Марка: {self.марка}")
        print(f"Модель: {self.модель}")
        print(f"Год выпуска: {self.год_выпуска}")
        print(f"Пробег: {self.пробег} км")
        print(f"--------------------------------")

    def сделать_поездку(self, км):
        """Увеличивает пробег на указанное количество километров."""
        if км > 0:
            self.пробег += км
            print(f"Автомобиль {self.марка} {self.модель} совершил поездку на {км} км.")
            print(f"Новый пробег: {self.пробег} км.")
        else:
            print("Расстояние для поездки должно быть положительным числом.")

# --- Демонстрация работы класса Автомобиль ---
print("Задание 1: Класс Автомобиль")

# Создаем несколько объектов класса Автомобиль
авто1 = Автомобиль("Toyota", "Camry", 2020, 50000)
авто2 = Автомобиль("Honda", "Civic", 2018, 75000)
авто3 = Автомобиль("BMW", "X5", 2022, 15000)

print("\nНачальная информация об автомобилях")
авто1.информация()
авто2.информация()
авто3.информация()

print("\nСовершаем поездки")
авто1.сделать_поездку(200)
авто2.сделать_поездку(1500)
авто3.сделать_поездку(50)
авто1.сделать_поездку(-10) # Пример некорректного ввода

print("\nИнформация об автомобилях после поездок")
авто1.информация()
авто2.информация()
авто3.информация()
```

```python
class Книга:
    общее_количество_созданных_книг = 0

    def __init__(self, название, автор, количество_страниц, количество_экземпляров):
        self.название = название
        self.автор = автор
        self.количество_страниц = количество_страниц
        self.количество_экземпляров = количество_экземпляров
        Книга.общее_количество_созданных_книг += 1 

    def информация(self):
        """Выводит полную информацию о книге."""
        print(f"Информация о книге")
        print(f"Название: {self.название}")
        print(f"Автор: {self.автор}")
        print(f"Количество страниц: {self.количество_страниц}")
        print(f"Доступно экземпляров: {self.количество_экземпляров}")
        print(f"Всего уникальных книг в системе: {Книга.общее_количество_созданных_книг}")
        print(f"--------------------------")

    def взять_книгу(self):
        """Уменьшает количество экземпляров на 1 при взятии книги."""
        if self.количество_экземпляров > 0:
            self.количество_экземпляров -= 1
            print(f"Книга '{self.название}' успешно взята. Осталось: {self.количество_экземпляров}")
        else:
            print(f"Книга '{self.название}' сейчас недоступна.")

    def вернуть_книгу(self):
        """Увеличивает количество экземпляров на 1 при возврате книги."""
        self.количество_экземпляров += 1
        print(f"Книга '{self.название}' успешно возвращена. Доступно: {self.количество_экземпляров}")
print("--- Задание 2: Класс Книга ---")

книга1 = Книга("Война и мир", "Лев Толстой", 1225, 5)
книга2 = Книга("Мастер и Маргарита", "Михаил Булгаков", 480, 3)
книга3 = Книга("1984", "Джордж Оруэлл", 328, 1)

print("\nНачальная информация о книгах")
книга1.информация()
книга2.информация()
книга3.информация() 

print("\nПробуем взять и вернуть книги")
книга1.взять_книгу()
книга1.взять_книгу()
книга2.взять_книгу()
книга3.взять_книгу()
книга3.взять_книгу()

print("\nИнформация о книгах после операций")
книга1.информация()
книга2.информация()
книга3.информация()

книга1.вернуть_книгу()
книга3.вернуть_книгу()

print("\nИнформация о книгах после возврата")
книга1.информация()
книга3.информация()

```

```python
import turtle

screen = turtle.Screen()
screen.setup(width=800, height=600) 
screen.bgcolor("lightblue") 
screen.title("Дом из Turtle")

pen = turtle.Turtle()
pen.speed(0) 
pen.pensize(3) 


def draw_square(x, y, size, color="brown"):
    """Рисует квадрат заданного размера и цвета."""
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.fillcolor(color)
    pen.begin_fill()
    for _ in range(4):
        pen.forward(size)
        pen.left(90)
    pen.end_fill()

def draw_rectangle(x, y, width, height, color="darkorange"):
    """Рисует прямоугольник заданных размеров и цвета."""
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.fillcolor(color)
    pen.begin_fill()
    pen.forward(width)
    pen.left(90)
    pen.forward(height)
    pen.left(90)
    pen.forward(width)
    pen.left(90)
    pen.forward(height)
    pen.left(90)
    pen.end_fill()

def draw_triangle(x1, y1, x2, y2, x3, y3, color="red"):
    """Рисует треугольник по трем координатам вершин."""
    pen.penup()
    pen.goto(x1, y1)
    pen.pendown()
    pen.fillcolor(color)
    pen.begin_fill()
    pen.goto(x2, y2)
    pen.goto(x3, y3)
    pen.goto(x1, y1) 
    pen.end_fill()

print("--- Задание 3: Рисунок дома с помощью Turtle ---")

house_body_size = 200
house_x = -100
house_y = -100
draw_square(house_x, house_y, house_body_size, "brown")

draw_triangle(-100, 100, 100, 100, 0, 180, "red")

window_size = 40
draw_square(-70, 30, window_size, "lightgray") 
draw_square(30, 30, window_size, "lightgray")  

door_width = 60
door_height = 100
draw_rectangle(-30, -100, door_width, door_height, "darkorange")

pen.penup()
pen.goto(20, -50) 
pen.pendown()
pen.dot(10, "black") 

pen.hideturtle()

screen.exitonclick()
print("--- Рисунок дома завершен. Нажмите на окно для закрытия. ---")
```
